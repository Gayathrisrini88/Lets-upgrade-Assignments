```python
#get the day of month, week number, day of year and day of week from a series of date strings
from dateutil.parser import parse
```

```python
date_series = pd.Series(['04 Jan 2010', '10-06-2014', '20110306', '2017/08/06', '2016-02-18', '2017-05-08T11:20'])
print("Original Series:")
print(date_series)
print()
date_series = date_series.map(lambda x: parse(x))
print("Day of month:")
print(date_series.dt.day.tolist())
print()
print("Day of year:")
print(date_series.dt.dayofyear.tolist())
print()
print("Week number:")
print(date_series.dt.weekofyear.tolist())
print()
print("Day of week:")
print(date_series.dt.dayofweek.tolist())
```

```python
#How to compute the euclidean distance between two series
x = pd.Series([1, 2, 3, 4, 5]) 
y = pd.Series([3, 6, 3, 2, 1]) 
p1 = np.sum([(a * a) for a in x]) 
p2 = np.sum([(b * b) for b in y]) 
# using zip() function to create an 
p3 = -1 * np.sum([(2 * a*b) for (a, b) in zip(x, y)]) 
dist = np.sqrt(np.sum(p1 + p2 + p3)) 
print("Series 1:", x) 
print("Series 2:", y) 
print("Euclidean distance between two series is:", dist) 
```

```python
#Matplotlib Assignment
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
%matplotlib inline
di=px.data.iris()
di.head()
```

```python
#Plot the frequency of the three species in the Iris Dataset.
freq=di['species'].value_counts()
print(freq);
freq.plot()
```

```python
#Plot the Pie of the three species in the Iris Dataset.
p=['virginica','versicolor','setosa']
q=di['species'].value_counts()
plt.pie(q,labels=p,autopct='%.2f%%',explode=[0.02,0.02,0.02])
plt.title('Species')
plt.legend()
plt.show()
```

```python
#Combining Box and Strip Plots for Species vs sepal length.
sns.catplot(x='species',y='sepal_length',kind='box',data=di)
```

```python
sns.stripplot(x='species',y='sepal_length',data=di)
```
